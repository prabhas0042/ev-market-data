```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)
plt.show()
```

```python


# Calculate correlation matrix
corr_matrix = df.pivot_table(index='year', columns='parameter', values='value').corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot time series for each parameter
parameters = df['parameter'].unique()
for param in parameters:
    subset = df[df['parameter'] == param]
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=subset, x='year', y='value', hue='powertrain', marker='o')
    plt.title(f'{param} Over Time')
    plt.xlabel('Year')
    plt.ylabel(param)
    plt.legend(title='Powertrain')
    plt.grid(True)
    plt.show()
```

```python
# Group by year and calculate aggregate statistics
yearly_summary = df.groupby('year').agg({'value': ['mean', 'sum', 'min', 'max']})
print(yearly_summary)
```

```python
# Use box plots to detect outliers for each parameter
for param in parameters:
    subset = df[df['parameter'] == param]
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=subset, x='year', y='value', hue='powertrain')
    plt.title(f'{param} Outlier Detection')
    plt.xlabel('Year')
    plt.ylabel(param)
    plt.legend(title='Powertrain')
    plt.show()
```

```python
# Create a pivot table to summarize the data
pivot_table = df.pivot_table(index='year', columns='parameter', values='value', aggfunc='sum')
print(pivot_table)
```

```python
#Function to plot time series data
def plot_time_series(df, parameter, title, ylabel, unit=''):
    plt.figure(figsize=(10, 6))
    for pt in df['powertrain'].unique():
        subset = df[(df['parameter'] == parameter) & (df['powertrain'] == pt)]
        plt.plot(subset['year'], subset['value'], marker='o', label=f'{pt} {unit}')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()
```

##Plot EV Sales Over Time

```python
plot_time_series(df, 'EV sales', 'EV Sales Over Time in India', 'Number of Vehicles', 'Vehicles')
```

##Plot EV Stock Over Time


```python
plot_time_series(df, 'EV stock', 'EV Stock Over Time in India', 'Number of Vehicles', 'Vehicles')
```

```python
plot_time_series(df, 'EV sales share', 'EV Sales Share Over Time in India', 'Percent', '%')
```

```python
plot_time_series(df, 'EV stock share', 'EV Stock Share Over Time in India', 'Percent', '%')
```
